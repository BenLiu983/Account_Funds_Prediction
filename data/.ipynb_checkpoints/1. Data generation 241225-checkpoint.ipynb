```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split

import lightgbm as lgb
from hyperopt import hp, tpe, fmin, Trials
```

```python
# Define the number of rows and columns
num_clients = 100000
num_features = 20

# Generate data for each column
np.random.seed(42)  # For reproducibility
data = {
    "bal_future": np.random.randint(0, 1000001, num_clients),  # Target column
    "bal": np.random.randint(0, 1000001, num_clients),
    "age": np.random.randint(18, 91, num_clients),  # Age range: 18-90
    "tenure": np.random.randint(0, 51, num_clients),  # Tenure range: 0-50 years
    "credit_score": np.random.randint(300, 851, num_clients),  # Credit score: 300-850
    "annual_income": np.random.randint(20000, 200001, num_clients),  # Income: $20k-$200k
    "mtg_balance": np.random.randint(0, 1000001, num_clients),  # Mortgage balance
    "credit_card_balance": np.random.randint(0, 50001, num_clients),  # Credit card balance
}

# Generate random data for columns 8 to 20
for i in range(8, num_features + 1):
    data[f"col{i}"] = np.random.randint(0, 1001, num_clients)

# Create the DataFrame
df = pd.DataFrame(data)

# Display a preview of the data
df.head()
```
